```python
import pandas as pd
import plotly.express as px

# Cargar los datos desde el archivo CSV
data_brent = pd.read_csv('brent_oil_prices.csv')

# Convertir la columna 'Date' a tipo datetime
data_brent['Date'] = pd.to_datetime(data_brent['Date'])

# Establecer la columna 'Date' como ndice del DataFrame
#data_brent.set_index('Date', inplace=True)

# Crear la grfica con plotly.express
fig = px.line(data_brent, x='Date', y='Close', 
              title='Evolucin Temporal del Precio de Cierre del Petrleo Brent', 
              labels={'Date': 'Fecha', 'Close': 'Precio de Cierre (USD)'},
              template='plotly_white')

# Configurar el hovermode para que sea 'x unified'
fig.update_layout(hovermode='x unified')

# Mostrar la grfica en Jupyter Notebook
fig.show()

# Guardar la figura como un archivo HTML
fig.write_html("precio_cierre_petroleo_brent.html")
```
